```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# set up better resolution for figures
# matplotlib
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

# seaborn
sb.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sb.set_context('notebook')
sb.set_style("ticks")
```

```python
metrics_hyperparameters = pd.read_csv("output_CCLE/metrics_hyperparameters.csv", index_col=[0])
metrics_hyperparameters
```

```python
metrics = metrics_hyperparameters[['ACCURACY', 'AUCROC', 'AUCPR']]
metrics
```

```python
metrics.describe()
```

```python
# make the data tidy for visualization
metrics_tidy = pd.melt(metrics, value_vars=list(metrics), var_name='metric', value_name='value')
metrics_tidy
```

```python
base_color = sb.color_palette()[0]
sb.barplot(data=metrics_tidy, x='metric', y='value', color=base_color);
#plt.xticks(rotation=15);
plt.xlabel('')
plt.ylabel('Performance of the RefRNN Model (CCLE)');
```
